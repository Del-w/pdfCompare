```python
import pdfplumber
import pytesseract
from pdf2image import convert_from_path
import cv2
import numpy as np

#Function to check if the PDF is text-based or scanned
def is_text_based_pdf(path, min_text_len=20):
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text and len(text.strip()) > min_text_len:
                return True
    return False

#Functions to extract text from a text-based PDF files
def extract_text_pdfplumber(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text.strip()

#Functions to extract text from scanned PDF files using OCR
def extract_text_ocr(path, dpi=300):
    images = convert_from_path(path, dpi=dpi)
    text = ""
    for image in images:
        img = np.array(image)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
        text += pytesseract.image_to_string(thresh) + "\n"
    return text.strip()

#Function to extract text from a PDF file, automatically determining the method based on the PDF type
def extract_text_auto(path):
    if is_text_based_pdf(path):
        print("Text-based PDF detected.")
        return extract_text_pdfplumber(path)
    else:
        print("Scanned PDF detected. Using OCR...")
        return extract_text_ocr(path)
```

```python
import pdfplumber
from pdf2image import convert_from_bytes
import pytesseract
import numpy as np
import cv2

#Funtion to extract text from a PDF file stream, automatically determining the method based on the PDF type
def extract_text(file_stream):
    text = ""
    #Check if the PDF is text-based or scanned
    try:
        with pdfplumber.open(file_stream) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"
    except:
        # If not text based, OCR fallback
        file_stream.seek(0)
        images = convert_from_bytes(file_stream.read())
        for image in images:
            img = np.array(image)
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
            text += pytesseract.image_to_string(thresh) + "\n"
    return text
```

```python
import spacy

nlp = spacy.load("en_core_web_lg")

def filter_unique_sentences(text1, text2, threshold=0.85):
    doc1 = nlp(text1)
    doc2 = nlp(text2)

    sents1 = list(doc1.sents)
    sents2 = list(doc2.sents)

    unique_sents2 = []

    for s2 in sents2:
        if all(s2.similarity(s1) < threshold for s1 in sents1):
            unique_sents2.append(s2.text.strip())

    # Optional: sort by sentence length or order from original
    unique_sents2 = sorted(unique_sents2, key=lambda x: sents2.index(next(s for s in sents2 if s.text.strip() == x)))

    # Combine with paragraph breaks for readability
    combined_text = text1.strip()
    if unique_sents2:
        combined_text += "\n\n--- Unique from PDF 2 ---\n\n" + "\n\n".join(unique_sents2)

    return combined_text
```
